```python
import pandas as pd
```

```python
df = pd.read_csv('/content/diamonds.csv')
df
```

```python
X = df.drop(['price'], axis=1)
y = df['price'].copy()
```

```python
X
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns
```

```python
cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

transformer = ColumnTransformer([
    ('cat', cat_pipeline, cat_features),
    ('num', num_pipeline, num_features)
])
```

```python
X_train_transformed = transformer.fit_transform(X_train)
X_train_transformed
```

```python
columns = transformer.get_feature_names_out()
```

```python
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=columns)
X_train_transformed_df
```

```python
X_test_transformed = transformer.transform(X_test)
X_test_transformed
```

```python
X_test_transformed_df = pd.DataFrame(X_test_transformed, columns=columns)
X_test_transformed_df
```

## DL

```python
import torch.nn as nn
import torch
from torch.utils.data import TensorDataset, DataLoader
```

```python
torch_train = torch.FloatTensor(X_train_transformed)
torch_test = torch.FloatTensor(X_test_transformed)
```

```python
y_train = torch.FloatTensor(y_train.to_numpy()).reshape(-1, 1)
y_test = torch.FloatTensor(y_test.to_numpy()).reshape(-1, 1)
```

```python
train_data = TensorDataset(torch_train, y_train)
test_data = TensorDataset(torch_train, y_test)
```

```python
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
test_loader = DataLoader(test_data, batch_size=32)
```

```python
n_features = X_train_transformed.shape[1]
```

```python
model = nn.Sequential(
    nn.Linear(in_features = n_features, out_features= 64),
    nn.ReLU(),
    nn.Linear(in_features = 64, out_features= 32),
    nn.ReLU(),
    nn.Linear(in_features = 32, out_features= 16),
    nn.ReLU(),
    nn.Linear(in_features = 16, out_features= 1),
)
```

```python
n_params = (9 + 1) * 64 + (64+ 1) * 32 + (32 + 1) * 16 + (16 + 1) * 1
n_params
```

```python
if torch.cuda.is_available():
  device = 'cuda'
else:
  device = 'cpu'
```

```python
def train(model, train_loader, optimizer, criterion, n_epochs):
  model.train()
  for epoch in range(n_epochs):
    total_loss = 0.
    for X_batch, y_batch in train_loader:
      X_batch, y_batch = X_batch.to(device), y_batch.to(device)
      y_pred = model(X_batch)
      loss = criterion(y_pred, y_batch)
      total_loss += loss.item()
      loss.backward()
      optimizer.step()
      optimizer.zero_grad()

    print(f'Epoch: {epoch+1}/{n_epochs}, Loss: {(total_loss / len(train_loader)) ** 0.5}')
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
mse = nn.MSELoss()
```

```python
model = model.to(device)
```

```python
train(model, train_loader, optimizer, mse, 20)
```
