```python
from PIL import Image
from torchvision import transforms
import torch
import torchvision
```

```python
cat1 = Image.open('/content/960px-Cat_November_2010-1a.jpg')
cat2 = Image.open('/content/CAT-CHAT-3.jpg')
dog = Image.open('/content/Dog_Breeds.jpg')
```

```python
weights = torchvision.models.ResNet34_Weights.IMAGENET1K_V1
model = torchvision.models.resnet34(weights)
```

```python
transform = weights.transforms()
```

```python
transform
```

```python
cat1_tensor = transform(cat1).unsqueeze(0)
cat2_tensor = transform(cat2).unsqueeze(0)
dog_tensor = transform(dog).unsqueeze(0)
```

```python
[name for name, param in model.named_parameters()]
```

```python
import torch.nn as nn
```

```python
feature_extractor = nn.Sequential(*list(model.children())[:-1])
```

```python
with torch.no_grad():
  emb1 = feature_extractor(cat1_tensor).view(-1)
  emb2 = feature_extractor(cat2_tensor).view(-1)
  emb3 = feature_extractor(dog_tensor).view(-1)
```

```python
import torch.nn.functional as F
```

```python
cat1cat2 = F.cosine_similarity(emb1, emb2, dim=0)
cat1dog = F.cosine_similarity(emb1, emb3, dim=0)
cat2dog = F.cosine_similarity(emb2, emb3, dim=0)
```

```python
print(f'Cat1 and Cat2 cosine similarity = {cat1cat2.item() * 100}')
print(f'Cat1 and Dog cosine similarity = {cat1dog.item() * 100}')
print(f'Cat2 and Dog cosine similarity = {cat2dog.item() * 100}')
```
